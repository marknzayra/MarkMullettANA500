```python
import xarray as xr
import pandas as pd
import seaborn as sns
```

```python
ds = xr.open_dataset('NOAAGlobalTemp_v5.1.0_gridded_s185001_e202307_c20230808T112655.nc')
```

```python
df = ds.to_dataframe()
```

```python
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df.reset_index(inplace=True)
```

```python
df.head()
```

I know the Idea was to use a Regression plot. It was taking more power that my computer could muster to plot each day. Since the data was from 1850 to present I went with the mean of each year. Yu can clearly see the trend for the data increasing over time.

```python


df.reset_index(inplace=True)


df['time'] = pd.to_datetime(df['time'], format='%Y-%m-%d')


df.set_index('time', inplace=True)


df_yearly = df.resample('Y').mean()


df_yearly.reset_index(inplace=True)


df_yearly['year'] = df_yearly['time'].dt.year


plt.figure(figsize=(12, 6))
sns.regplot(x='year', y='anom', data=df_yearly, scatter_kws={'s': 10}, color='b')


plt.xlabel('Year')
plt.ylabel('Yearly Average Temperature Anomaly (°C)')
plt.title('Regression Plot of Ocean Temperature Anomalies (Yearly Averages)')


plt.grid(True)
plt.show()

```

I like the way this one looks as well. You can clearly see the changes over time.

```python
df.reset_index(inplace=True)

df['time'] = pd.to_datetime(df['time'], format='%Y-%m-%d')

df.set_index('time', inplace=True)

df_yearly = df.resample('Y').mean()

df_yearly.reset_index(inplace=True)

df_yearly['year'] = df_yearly['time'].dt.year

plt.figure(figsize=(12, 6))
plt.plot(df_yearly['year'], df_yearly['anom'], label='Temperature Anomaly (°C)')


plt.xlabel('Time')
plt.ylabel('Temperature Anomaly (°C)')
plt.title('Global Ocean Temperature Anomalies (1850-2018)')


plt.grid(True)
plt.legend()
plt.show()
```
